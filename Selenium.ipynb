```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# Set up the Chrome driver
service = Service(executable_path="/content/chromedriver.exe")
driver = webdriver.Chrome(service=service)

# Open Bing
driver.get("https://www.bing.com/")

# Wait for the page to load
time.sleep()

# Find the search box using its name attribute
box = driver.find_element(By.NAME, "q")

# Send the search query and press Enter
box.send_keys("Final")
box.send_keys(Keys.RETURN)

# Optional: wait to observe the result
time.sleep(10)

# Close the browser
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import os # Import the os module

# Set up the Chrome driver
# Grant execute permissions to the chromedriver executable
chromedriver_path = "/content/chromedriver.exe"
if os.path.exists(chromedriver_path):
    os.chmod(chromedriver_path, 0o755) # Grant read, write, and execute permissions for the owner, and read and execute for others

service = Service(executable_path=chromedriver_path)
driver = webdriver.Chrome(service=service)

# Open Bing
driver.get("https://www.bing.com/")

# Wait for the page to load
# The time.sleep() function requires an argument for the duration in seconds.
# Adding a small delay, e.g., 2 seconds, to allow the page to load.
time.sleep(2)

# Find the search box using its name attribute
box = driver.find_element(By.NAME, "q")

# Send the search query and press Enter
box.send_keys("Final")
box.send_keys(Keys.RETURN)

# Optional: wait to observe the result
time.sleep(10)

# Close the browser
driver.quit()
```
