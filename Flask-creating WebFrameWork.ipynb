```python
!pip install flask-ngrok

# Import necessary libraries
from flask import Flask, jsonify, request  # Flask framework and tools for handling requests and JSON responses
import threading  # For running the Flask app in the background (so notebook doesn’t freeze)

# Create a Flask web application instance
app = Flask(__name__)

# Define the home route (GET request to "/")
@app.route('/')
def home():
    # Returns a simple welcome message when someone accesses the root URL
    return "Welcome to my Flask API running!"

# Define a GET API endpoint to greet a user
@app.route('/api/greet', methods=['GET'])
def greet():
    # Get the 'name' query parameter from the URL; default to "Guest" if not provided
    name = request.args.get('name', 'Guest')
    # Return a JSON response with a greeting message
    return jsonify({"message": f"Hello, {name}!"})

# Define a POST API endpoint to calculate the sum of two numbers
@app.route('/api/sum', methods=['POST'])
def add():
    # Parse JSON data from the request body
    data = request.get_json(force=True)
    try:
        # Get 'a' and 'b' from JSON, convert to float, use 0 as default
        a = float(data.get('a', 0))
        b = float(data.get('b', 0))
        # Return the result as a JSON response
        return jsonify({"result": a + b})
    except:
        # If input is invalid (e.g., strings instead of numbers), return an error
        return jsonify({"error": "Invalid input"}), 400

# Define a function to start the Flask app
def run_app():
    # Start the Flask app on port 5000 (default localhost)
    app.run(port=5000)

# Run the Flask app in a separate thread so it doesn't block the notebook
thread = threading.Thread(target=run_app)
thread.start()
```
