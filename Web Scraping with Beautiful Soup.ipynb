```python
#Import Libraries

!pip install requests
!pip install beautifulsoup4
import requests # import the requests library
from bs4 import BeautifulSoup # import BeautifulSoup from bs4(Module)

#Send an HTTP Request to the Website

url = 'https://example.com'
response = requests.get(url) #requesrt for url

# Optional: Check response status
if response.status_code == 200:
    print("Success")
else:
    print("Failed to retrieve the page")

#Parse the HTML Content
soup = BeautifulSoup(response.text, 'html.parser')


#'html.parser' is Python’s built-in parser.
#Other options: 'lxml', 'html5lib' (must install separately).
```

```python
#Navigate and Search the HTML Tree
#You can search by tag name, class, id, attributes, etc.

#Find by Tag
title = soup.title
print(title.text)
```

```python
#Find First Match

first_paragraph = soup.find('p')
print(first_paragraph.text)
```

```python
#Find All Matches

all_links = soup.find_all('a')
for link in all_links:
    print(link['href'])
```

```python
all_links = soup.find_all('a')

for link in all_links:
    href = link.get('href')
    if href:
        print(href)
```

```python
#Example Use Case

import requests
from bs4 import BeautifulSoup

url = 'https://quotes.toscrape.com/'
response = requests.get(url)

soup = BeautifulSoup(response.text, 'html.parser')

quotes = soup.find_all('span', class_='text')
authors = soup.find_all('small', class_='author')

for quote, author in zip(quotes, authors):
    print(f"{quote.text} — {author.text}")
```
