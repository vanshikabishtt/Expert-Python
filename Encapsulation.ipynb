```python
class Laptop:
    def __init__(self,price):
        self.__price = price # Private attribute

    def set_price(self, new_price):
        self.__price = new_price  # Update the price

    def get_price(self):
        return self.__price  #return the price


laptop = Laptop(50000)
laptop.set_price(45000)  # Set the price
print(laptop.get_price())  # Get the price
```

```python
class BankAccount:
    def __init__(self,balance):
        self.__balance = balance  # Private attribute

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount  # Increase balance

    def withdraw(self, amount):
        if amount > 0 and amount <= self.__balance:
            self.__balance -= amount  # Decrease balance

    def check_balance(self):
        return self.__balance  # Return current balance


account = BankAccount(70000)
account.deposit(500)  # Deposit money
print(account.check_balance())  # Check balance

account.withdraw(200)  # Withdraw money
print(account.check_balance())  # Check balance

account.withdraw(400)  # Try to withdraw more than available balance
```

```python
class Movie:
    def __init__(self):
        self.__rating = 0  # Private attribute

    def set_rating(self, rating):
        if 1 <= rating <= 10:
            self.__rating = rating  # Set the rating
        else:
            print("Rating must be between 1 and 10.")

    def get_rating(self):
        return self.__rating  # Return the rating

movie = Movie()
movie.set_rating(8)  # Set a valid rating
print(movie.get_rating())  # Get the rating

movie.set_rating(12)  # Try to set an invalid rating

```

```python
class Book:
    def __init__(self, title, author):
        self.__title = title  # Private attribute for title
        self.__author = author  # Private attribute for author

    def get_title(self):
        return self.__title  # Return the title

    def get_author(self):
        return self.__author  # Return the author

book = Book("To Kill a Mockingbird", "Harper Lee")
print(book.get_title())   # Get the title
print(book.get_author())  # Get the author

```

```python
class Car:
    def __init__(self,speed):
        self.__speed = speed  # Private attribute for speed

    def increase_speed(self, amount):
        self.__speed += amount  # Increase speed

    def decrease_speed(self, amount):
        self.__speed -= amount
        if self.__speed < 0:
            self.__speed = 0  # Ensure speed doesn't go below 0

    def get_speed(self):
        return self.__speed  # Return the current speed

car = Car(int(input("enter speed")))
car.increase_speed(int(input("enter speed")))  # Increase speed by 30
print(car.get_speed())   # Get the current speed

car.decrease_speed(int(input("enter speed")))  # Decrease speed by 10
print(car.get_speed())   # Get the current speed


```
